```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr
```

```python
outpath = "output/Validation_TS/"
```

```python
datanames = ["TSP1_10x_with_postprocessing_nopanc_cellann_sharedjuncs_sharedonts_10x","TSP2_10x_rerun_with_postprocessing_3prime_cellann_sharedjuncs_sharedonts_10x"]
dfs = []

z_cols = ["scZ"]#,"svd_z0","svd_z1","svd_z2"]
sig_genes = []
for dataname in datanames:
  print(dataname)
  df = pd.read_parquet("../data/{}_sym_SVD_normdonor_S_0.1_z_0.0_b_5.pq".format(dataname),columns=["ontology","geneR1A_uniq","scZ","svd_z0","svd_z1","svd_z2","cell_gene"])
  sig = pd.read_csv("/scratch/PI/horence/JuliaO/single_cell/SZS_pipeline2/scripts/output/variance_adjusted_permutations/{}_pvals_ontology-dummy_100_S_0.1_z_0.0_b_5_sig.tsv".format(dataname),sep="\t")
  sig_genes.append(set(sig[sig["sig"]]["geneR1A_uniq"].unique()))
  df = df.drop_duplicates("cell_gene")
  df["ontology_gene"] = df["ontology"] + df["geneR1A_uniq"]
  df["num_ont_gene"] = df["ontology_gene"].map(df.groupby("ontology_gene")["cell_gene"].nunique())
  df = df[df["num_ont_gene"] > 10]
  for z_col in z_cols:
    df["median_" + z_col] = df["ontology_gene"].map(df.groupby("ontology_gene")[z_col].median())
  df.drop_duplicates("ontology_gene")
  dfs.append(df)
```

```python
both_sig = set.intersection(*sig_genes)
for i in range(len(dfs)):
  dfs[i] = dfs[i].drop_duplicates("ontology_gene")
merged = dfs[0].merge(dfs[1],on="ontology_gene",suffixes=["_TSP1","_TSP2"])
```

```python

zcol_dict = {"scZ" : "SpliZ", "svd_z0" : "SpliZVD"}
for z_col in ["scZ"]:
#   plt.plot(merged["median_{}_TSP1".format(z_col)],merged["median_{}_TSP2".format(z_col)],linestyle="",marker="o",alpha = 0.4)
  plt.hexbin(merged["median_{}_TSP1".format(z_col)],merged["median_{}_TSP2".format(z_col)],gridsize=30,cmap="Greys",bins="log")

  plt.title("{}\npearson: {:.3f} spearman: {:.3f}".format(zcol_dict[z_col],merged["median_{}_TSP1".format(z_col)].corr(merged["median_{}_TSP2".format(z_col)],method="pearson"),merged["median_{}_TSP1".format(z_col)].corr(merged["median_{}_TSP2".format(z_col)],method="spearman")))
  
  plt.xlabel("median {} for individual 1 10x".format(zcol_dict[z_col]))
  plt.ylabel("median {} for individual 2 10x ".format(zcol_dict[z_col]))
  plt.savefig("{}{}_corr_all.png".format(outpath,z_col),bbox_inches="tight")
  plt.show()
  
  sub_df = merged[merged["geneR1A_uniq_TSP1"].isin(both_sig)]
  
  plt.hexbin(sub_df["median_{}_TSP1".format(z_col)],sub_df["median_{}_TSP2".format(z_col)],gridsize=30,cmap="Greys",bins="log")

  plt.title("genes called in both 10x datasets by the {}\npearson: {:.3f} spearman: {:.3f}".format(zcol_dict[z_col],sub_df["median_{}_TSP1".format(z_col)].corr(sub_df["median_{}_TSP2".format(z_col)],method="pearson"),sub_df["median_{}_TSP1".format(z_col)].corr(sub_df["median_{}_TSP2".format(z_col)],method="spearman")))
  plt.xlabel("median {} for individual 1".format(zcol_dict[z_col]))
  plt.ylabel("median {} for individual 2".format(zcol_dict[z_col]))
  plt.savefig("{}{}_corr_sig.png".format(outpath,z_col),bbox_inches="tight")

  plt.show()
  print(pearsonr(sub_df["median_{}_TSP2".format(z_col)],sub_df["median_{}_TSP1".format(z_col)]))
```
